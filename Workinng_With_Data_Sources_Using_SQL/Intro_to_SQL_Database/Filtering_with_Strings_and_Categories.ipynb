#### Finding Unique Values with SELECT DISTINCT

```python
import sqlite3
conn = sqlite3.connect(r"C:\Users\ajaya\Downloads\superstore.db")
cur = conn.cursor()
```

```python
cur.execute('''SELECT DISTINCT order_id,customer_name
from orders 
where city = 'Buffalo'
and state = 'New York'
''')
for row in cur:
    print(row)
```

```python
cur.execute('''SELECT DISTINCT ship_mode,state 
from orders
where state IN ('District of Columbia','North Dakota','Vermont','West Virginia')
''')

for row in cur:
    print(row)

                
```

```python
cur.execute('''SELECT order_id, city, state, sales
  FROM orders
 WHERE state NOT IN ('California', 'Texas', 'New York')
       AND sales > 5000;
''')
for row in cur:
    print(row)
```

### Searching for Text Patterns with LIKE (Part 1)

```python
cur.execute('''select DISTINCT subcategory,product_name from orders
where product_name like '%keyboard%'
''')

for row in cur:
    print(row)
```

### Searching for Text Patterns with LIKE (Part 2)

```python
cur.execute('''
SELECT DISTINCT product_name
  FROM orders
 WHERE product_name LIKE 'Pr___ C_l____ %';''')

for row in cur:
    print(row)
```

```python
cur.execute('''select product_name,subcategory,sales,profit
from orders 
where (sales > 5000 and profit > 1000)
OR 
(profit <-10 and sales < 10) LIMIT 10;
''')

for row in cur:
    print(row)
```
