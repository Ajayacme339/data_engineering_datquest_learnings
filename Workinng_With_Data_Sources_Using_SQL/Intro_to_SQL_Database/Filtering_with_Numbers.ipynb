### Basic Comparison Operators

```python
import sqlite3
conn = sqlite3.connect(r"C:\Users\ajaya\Downloads\superstore.db")
cur = conn.cursor()
```

```python
cur.execute('''select order_id,product_name,sales,discount,profit 
from orders 
where profit < -1000''')
for row in cur:
    print(row)
```

#### Order of Execution

```python
cur.execute('''select order_id,subcategory,product_name,
sales/quantity AS price_per_unit from 
orders 
where sales/quantity < 0.50;''')
for row in cur:
    print(row)
```

```python
#### Comparison Operators Continued
cur.execute("select * from orders where customer_name = 'Sean Miller'")
for row in cur:
    print(row)
```

#### Checking for Membership in a Consecutive Range

```python
### Checking for Missing Values
cur.execute('''select * from orders 
where segment IS NULL''')
 
for row in cur:
    print(row)
```

```python
cur.execute('''select product_name,sales/quantity AS price_per_unit
from orders
where subcategory = 'Storage' 
and  price_per_unit > 300;''')

for row in cur:
    print(row)
```

```python
cur.execute('''SELECT product_name,profit,quantity
from orders 
where profit < 0 OR quantity = 1
LIMIT 20;''')

for row in cur:
    print(row)
```
