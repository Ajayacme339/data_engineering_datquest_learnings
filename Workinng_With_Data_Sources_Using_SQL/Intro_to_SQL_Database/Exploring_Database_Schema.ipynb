```python

import sqlite3
conn = sqlite3.connect(r"C:\Users\ajaya\Downloads\superstore.db")
cur = conn.cursor()
response = cur.execute('''
SELECT order_id, product_name, sales, quantity,
       CASE 
       WHEN sales BETWEEN 0 AND 50 THEN 'small sale'
       WHEN sales BETWEEN 51 AND 100 THEN 'medium sale'
       ELSE 'large sale'
       END AS sales_amount                        
  FROM orders
 WHERE order_id LIKE 'CA%'
 ORDER BY quantity
 LIMIT 10;
 ''')
for all in response:
    print(all)

```

### Reading from table

```python
result = cur.execute("SELECT * FROM orders LIMIT 5")
for row in result:
    print(row)
```

```python
result = cur.execute("SELECT * FROM returns LIMIT 4;")
for row in result:
    print(row)
```

### Select Rows and Columns

```python
cur.execute('''SELECT order_date,
order_id,
product_name,
sales,
quantity from orders LIMIT 5;
''')
for row in cur:
    print(row)
```

### Defining Relational Databases

```python

cur.execute('''select city,state,region 
from orders LIMIT 10;''')
for row in cur:
    print(row)
```

### Aliasing the result

```python
cur.execute('''SELECT
city AS City,state AS State,postal_code AS 'ZIP Code'
from orders LIMIT 10''')

for row in cur:
    print(row)
```
