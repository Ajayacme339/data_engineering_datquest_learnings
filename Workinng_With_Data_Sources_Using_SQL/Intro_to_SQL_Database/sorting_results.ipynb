#### Sorting, Ascending vs. Descending

```python
import sqlite3
conn = sqlite3.connect(r"C:\Users\ajaya\Downloads\superstore.db")
cur = conn.cursor()
```

```python
cur.execute('''SELECT order_id,product_name,profit 
from orders 
ORDER BY profit DESC LIMIT 20''')

for row in cur:
    print(row)
```

```python
### Ordering by Multiple Fields
cur.execute('''SELECT product_name, sales, discount, profit
  FROM orders
 ORDER BY discount DESC, profit LIMIT 20;''')

for row in cur:
    print(row)
```

### Ordering by Ordinal

```python
cur.execute('''SELECT ship_date, ship_mode, city, state, region
  FROM orders
 ORDER BY 1 DESC, 3 LIMIT 20;''')

for row in cur:
    print(row)
```

### ORDER BY with WHERE

```python
cur.execute('''select subcategory,sales,profit 
from orders 
where category = 'Furniture' 
ORDER BY sales DESC,profit ASC LIMIT 20;''')

for row in cur.fetchall():
    print(row)
```

### ORDER BY vs LIMIT

```python
cur.execute('''SELECT order_id, quantity
  FROM orders
 WHERE region = 'Central' AND category = 'Office Supplies'
 ORDER BY quantity DESC
 LIMIT 10;''')

for row in cur:
    print(row)
```

```python
cur.execute('''SELECT product_name, sales/quantity AS price_per_unit
  FROM orders
 WHERE sales/quantity < 95
 ORDER BY price_per_unit DESC
 LIMIT 10;''')

for row in cur:
    print(row)
```

```python
### Coding with Style
cur.execute('''SELECT subcategory, product_name, quantity
  FROM orders
 WHERE category = 'Furniture'
 ORDER BY product_name
 LIMIT 10;''')

for row in cur:
    print(row)
```
