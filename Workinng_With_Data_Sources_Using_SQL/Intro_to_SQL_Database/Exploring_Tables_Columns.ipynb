```python
import sqlite3
conn = sqlite3.connect(r"C:\Users\ajaya\Downloads\superstore.db")
cur = conn.cursor()
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cur.fetchall()
for table in tables:
    print(table[0])
```

### Understanding Data Types

```python
cur.execute("select sales ,discount,profit from orders LIMIT 5")
for row in cur:
    print(row)
```

```python
####Table Information with PRAGMA
cur.execute("PRAGMA table_info(orders);")
for row in cur:
    print(row)
#print(cur.fetchall())
```

#### Operations on Numeric Columns

```python
cur.execute('''select order_id, sales, profit,profit/sales AS profit_margin
FROM orders 
LIMIT 8;''')

for row in cur:
    print(row)
```

 ### Integer Division

```python
cur.execute('''SELECT product_id,quantity,CAST(quantity AS REAL)/2  AS even_or_odd
from orders
LIMIT 5;''')
for row in cur:
    print(row)
```

### Functions on Fields: ROUND()

```python
cur.execute('''SELECT  order_id,
sales,quantity,ROUND(sales/quantity) AS price_per_unit
from orders 
LIMIT 10;''')
for row in cur:
    print(row)
```

### Functions on Fields: UPPER() and LOWER()

```python
cur.execute('''select LOWER(customer_name) AS customer_name_lower from orders
LIMIT 10;''')
for row in cur:
    print(row)
```

#### Concatenating Fields

```python
cur.execute('''SELECT order_id, 
       region, 
       state, 
       "Superstore " || city AS local_store
  FROM orders
 LIMIT 10;''')
 
for row in cur:
    print(row)
```

### Constant Value 

```python
cur.execute("select *,51000 from  managers;")
for row in cur.fetchall():
    print(row)
```

```python
cur.execute('''SELECT city || ", " || state || " " || postal_code AS address,
	   sales,
       ROUND(sales + 4.99 + sales*0.07, 2) AS total_cost,
	   ROUND(sales * 0.07, 2) AS tax,
	   4.99 AS shipping_cost
  FROM orders
 LIMIT 10;''')

for row in cur:
    print(row)
```
