### Binary and Positional Number Systems

```python
n =  2**32
print(n)
```

```python
weight_digit_2 = 10**2
value_digit_2 = 200
weight_digit_5 = 10**3
value_digit_5 = 5000
print(weight_digit_2)
print(value_digit_2)
print(weight_digit_5)
print(value_digit_5)
```

```python
decimal_1 = 25
decimal_2 = 14
```

```python
bin_1 = bin(12345)
bin_2 = bin(1337)
print(bin_1)
print(bin_2)
```

```python
base_8_to_10 = int('435',8)
base_7_to_10 = int('10',7)
print(base_8_to_10)
print(base_7_to_10)
```

```python
hex_3501 = hex(3501)
decimal_F = int('F',16)
print(hex_3501)
print(decimal_F)
```

```python
red_hex = hex(213)
print(red_hex)
green_hex = hex(111)
print(green_hex)
blue_hex = hex(56)
print(blue_hex)
```

```python
octal_999 = oct(999)
original = int(octal_999,8)
print(original)
```
